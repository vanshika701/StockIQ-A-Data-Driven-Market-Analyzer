```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stocks=pd.read_csv('stock_data_july_2025.csv')
```

```python
stocks.head()
```

```python
stocks.info()
```

```python
stocks['Date']=pd.to_datetime(stocks['Date'])
```

```python
stocks.info()
```

```python
stocks.describe()
```

```python
stocks.shape
```

```python
plt.figure(figsize=(20, 15))
sns.displot(stocks['Open Price'],kde=True,bins=80)
plt.show()
```

#### In this the data is skewed towards the lesser opening price, which means that the opening price of most of the stocks are less while there are some priced stocks ranging bw 3500-4000

```python
plt.figure(figsize=(20, 15))
sns.displot(stocks['Close Price'],kde=True,bins=80)
plt.show()
```

#### The closing price plot shows similar variations to that of the opening price plot, which means most of the stocks dont show much variation in a day

```python
sns.scatterplot(x='Open Price', y='Close Price', data=stocks)
```

#### This scatterplot suggests that intraday stock variation is low

```python
stocks['Daily Return']=((stocks['Close Price']-stocks['Open Price'])/stocks['Open Price'])*100
```

```python
stocks.head()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(x='Date', y='Daily Return', data=stocks)
plt.xticks(rotation=45)
plt.title('Daily Return Over Time')
plt.show()
```

```python
sns.displot(stocks['Daily Return'],kde=True,bins=80)
```

```python
stocks.isnull().sum()
```

```python
sns.heatmap(stocks.isnull())
```

### Indicates that there are no missing values in the dataset

```python
plt.figure(figsize=(15, 10))

# Create scatter plot
sns.scatterplot(data=stocks, x='EPS', y='PE Ratio', hue='Sector', alpha=0.5, s=80)


# Add title and labels
plt.title('PE Ratio vs EPS by Sector',fontsize=20)
plt.xlabel('Earnings Per Share (EPS)',fontsize=16)
plt.ylabel('PE Ratio',fontsize=16)
plt.legend(title='Sector',title_fontsize=16, fontsize=14, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

### the plot clearly indicates the Industrial and Consumer discretionary sectors experience the highest EPS with a moderate to low PE Ratio, which means the stock is undervalued and is very beneficial.(These sectors are highly beneficial) 
### And the technology sector experiences high PE and low EPS, meaning they're overvalued and only investors applying would be interested in long term inverstments.


```python
plt.figure(figsize=(12, 7))

sns.scatterplot(
    data=stocks,
    x='EPS',
    y='PE Ratio',
    hue='Sector',
    size='Market Cap',           # Bubble size
    sizes=(20, 300),             # Adjust min/max size
    alpha=0.6,
    palette='tab10'
)

plt.title('PE Ratio vs EPS (Bubble Size = Market Cap)', fontsize=16)
plt.xlabel('Earnings Per Share (EPS)', fontsize=14)
plt.ylabel('PE Ratio', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Sector', fontsize=10, title_fontsize=12)

plt.tight_layout()
plt.show()
```
