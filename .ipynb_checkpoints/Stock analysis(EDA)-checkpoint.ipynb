```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stocks=pd.read_csv('stock_data_july_2025.csv')
```

```python
stocks.head()
```

```python
stocks.info()
```

```python
stocks.describe()
```

```python
stocks.shape
```

```python
plt.figure(figsize=(20, 15))
sns.displot(stocks['Open Price'],kde=True,bins=80)
plt.show()
```

#### In this the data is skewed towards the lesser opening price, which means that the opening price of most of the stocks are less while there are some priced stocks ranging bw 3500-4000

```python
plt.figure(figsize=(20, 15))
sns.displot(stocks['Close Price'],kde=True,bins=80)
plt.show()
```

#### The closing price plot shows similar variations to that of the opening price plot, which means most of the stocks dont show much variation in a day

```python
sns.scatterplot(x='Open Price', y='Close Price', data=stocks)
```

#### This scatterplot suggests that intraday stock variation is low

```python
stocks['Daily Return']=((stocks['Close Price']-stocks['Open Price'])/stocks['Open Price'])*100
```

```python
stocks.head()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(x='Date', y='Daily Return', data=stocks)
plt.xticks(rotation=45)
plt.title('Daily Return Over Time')
plt.show()
```

```python
sns.displot(stocks['Daily Return'],kde=True,bins=80)
```

```python
plt.figure(figsize=(15, 10))

# Create scatter plot
sns.scatterplot(data=stocks, x='EPS', y='PE Ratio', hue='Sector', alpha=0.5, s=80)


# Add title and labels
plt.title('PE Ratio vs EPS by Sector',fontsize=20)
plt.xlabel('Earnings Per Share (EPS)',fontsize=16)
plt.ylabel('PE Ratio',fontsize=16)
plt.legend(title='Sector',title_fontsize=16, fontsize=14, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

### the plot clearly indicates the Industrial and Consumer discretionary sectors experience the highest EPS with a moderate to low PE Ratio, which means the stock is undervalued and is very beneficial.(These sectors are highly beneficial) 
### And the technology sector experiences high PE and low EPS, meaning they're overvalued and only investors applying would be interested in long term inverstments.


```python
plt.figure(figsize=(12, 7))

sns.scatterplot(
    data=stocks,
    x='EPS',
    y='PE Ratio',
    hue='Sector',
    size='Market Cap',           # Bubble size
    sizes=(20, 300),             # Adjust min/max size
    alpha=0.6,
    palette='tab10'
)

plt.title('PE Ratio vs EPS (Bubble Size = Market Cap)', fontsize=16)
plt.xlabel('Earnings Per Share (EPS)', fontsize=14)
plt.ylabel('PE Ratio', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Sector', fontsize=10, title_fontsize=12)

plt.tight_layout()
plt.show()
```

```python
melted = pd.melt(
    stocks,
    id_vars=['Sector'],
    value_vars=['52 Week High', '52 Week Low'],
    var_name='Price Type',   # This will be "52 Week High" or "52 Week Low"
    value_name='Price'       # This holds the actual price values
)
```

```python
melted
```

```python
plt.figure(figsize=(18,14))
sns.barplot(x='Sector',y='Price',data=melted, hue='Price Type')
plt.tight_layout()
plt.xticks(rotation=45)  # Rotate x labels if needed
plt.title('Price Comparison by Sector',fontsize=20)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Price Type', fontsize=20, title_fontsize=22)
plt.tick_params(labelsize=20)
plt.xlabel('Sector', fontsize=24)
plt.ylabel('Price', fontsize=24)
```

### Consumer Discretionary and Tech have seen large price movements — likely due to speculative buying or growth cycles — while Utilities and Consumer Staples have remained stable, appealing to risk-averse investors.
### Also the difference between both the high and low prices represents volarilty of the stocks in the given sector.

## Making a heatmap to analyse 

```python
stocks.columns
```

```python
num_cols = stocks.select_dtypes(include=['number'])
```

```python
num_cols
```

```python
corr_matrix = num_cols.corr()
```

```python
corr_matrix
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_matrix,
    annot=True,       # show correlation values
    fmt=".2f",        # format numbers to 2 decimal places
    cmap='PiYG',  # color scheme: blue = negative, red = positive
    square=True,
    linewidths=0.5
)

plt.title("Correlation Heatmap of Stock Features", fontsize=16)
plt.tight_layout()
plt.show()
```

#### Our heatmap revealed strong positive correlation between stock price features and earnings per share, highlighting profitability as a key price driver. Dividend yield, conversely, showed a negative relationship with both PE ratio and price metrics, signaling investor preference toward growth over income in high-performing sectors.
